```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#let's make synthetic datapoints
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
## prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cr=classification_report(y_test,y_pred)
print("Classification Report \n: ",cr)
cm=confusion_matrix(y_test,y_pred)
print("Confustion Matrix \n: ",cm)
```

```python
#using rbf kernel
rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred1=svc.predict(X_test)
cr=classification_report(y_test,y_pred1)
print("Classification Report \n: ",cr)
cm=confusion_matrix(y_test,y_pred1)
print("Confustion Matrix \n: ",cm)
```

```python
#using polynomial kernel
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
y_pred2=polynomial.predict(X_test)
cr=classification_report(y_test,y_pred2)
print("Classification Report \n: ",cr)
cm=confusion_matrix(y_test,y_pred2)
print("Confustion Matrix \n: ",cm)
```

```python
#using sigmoid kernel
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_pred3=sigmoid.predict(X_test)
cr=classification_report(y_test,y_pred3)
print("Classification Report \n: ",cr)
cm=confusion_matrix(y_test,y_pred3)
print("Confustion Matrix \n: ",cm)
```

```python
## hyperparameter tuning using gridsearchcv

from sklearn.model_selection import GridSearchCV

## defining parameter range
param_grid={
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(estimator=SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3) ## refit parameter for every validation it will refit again
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(X_test)
cr=classification_report(y_test,y_pred4)
print("Classification Report \n: ",cr)
cm=confusion_matrix(y_test,y_pred4)
print("Confustion Matrix \n: ",cm)
```
