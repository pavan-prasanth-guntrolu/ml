```python
## Feature encoding (Label Encoding And Onehot Encoding)
## It is important that we should we feature encoding only after train_test_split
df.columns
```

```python
#independent and dependent features
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
## feature encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train["sex"])
X_train['smoker']=le2.fit_transform(X_train["smoker"])
X_train['time']=le3.fit_transform(X_train["time"])
```

```python
X_train
```

```python
X_test['sex']=le1.transform(X_test["sex"])
X_test['smoker']=le2.transform(X_test["smoker"])
X_test['time']=le3.transform(X_test["time"])
```

```python
## OneHotEncoding
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop="first"),[3])],remainder="passthrough")
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
X_train
```

```python
## svr - support vector regression
```

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print("r2_score",r2_score(y_test,y_pred))
print("Mean absolute error",mean_absolute_error(y_test,y_pred))
```

```python
## hyperparameter tuning using GridSearchCV
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid={
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(estimator=SVR(),param_grid=param_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_predict=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print("r2_score",r2_score(y_test,grid_predict))
print("Mean absolute error",mean_absolute_error(y_test,grid_predict))
```
