```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("flight-price.xlsx")
df.head()
```

```python
df["day"]=df["Date_of_Journey"].str.split("/").str[0]
df["month"]=df["Date_of_Journey"].str.split("/").str[1]
df["year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df["day"]=df["day"].astype(int)
df["month"]=df["month"].astype(int)
df["year"]=df["year"].astype(int)
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.tail()
```

```python
df["Arrival_Hours"]=df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Minutes"]=df["Arrival_Time"].str.split(":").str[1].str.split(" ").str[0]
```

```python
df.head(2)
```

```python
df["Dep_Hour"]=df["Dep_Time"].str.split(":").str[0]
df["Dep_Min"]=df["Dep_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
df["Arrival_Hours"]=df["Arrival_Hours"].astype(int)
df["Arrival_Minutes"]=df["Arrival_Minutes"].astype(int)
df["Dep_Hour"]=df["Dep_Hour"].astype(int)
df["Dep_Min"]=df["Dep_Min"].astype(int)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.drop("Dep_Time",inplace=True,axis=1)
df.drop("Arrival_Time",inplace=True,axis=1)
```

```python
df.head(2)
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].mode()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head(2)
```

```python
df.drop("Route",inplace=True,axis=1)
```

```python
df.head(2)
```

```python
df["Dur_Hour"]=df["Duration"].str.split(" ").str[0].str.split("h").str[0].str.split("m").str[0]
df["Dur_Min"]=df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
df["Dur_Min"]=df["Dur_Min"].fillna(0)
```

```python
df["Dur_Min"]=df["Dur_Min"].astype(int)
df["Dur_Hour"]=df["Dur_Hour"].astype(int)
df.info()
```

```python
df.drop("Duration",inplace=True,axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder()

encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray()

```

```python
encoded=pd.DataFrame(encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df,encoded],axis=1)
```

```python
df.head()
```

```python
df.drop(["Airline","Source","Destination"],axis=1,inplace=True)
```
